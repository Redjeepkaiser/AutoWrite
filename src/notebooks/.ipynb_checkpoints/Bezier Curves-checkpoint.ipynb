# Introduction

In this project we improve handwriting. To do this the data used to represent the handwriting first needs to be transformed to another representation that is more memmory efficient, filters out outliers and is a better representation for RNNs to work with, this is further discussed in the thesis. In this notebook it will be shown how strokes (a list of coordinates with timestamps), which is the initial representation of handwriting, can be represented as Bezier curves. This representation can then be used to further analyze handwriting.

# Raw Touch Points

For reading the data from a csv file we will use pandas.

```python
import pandas as pd
pd.__version__
```

Autowrite is the program used to record input and it saves the recorded input (as strokes) in "src/autowrite/output/strokes.csv", so we will read some example input from there. We need to convert the strings from the csv file back into tuples.

```python
import ast
```

```python
input_location = "./../autowrite/output/strokes.csv"

strokes = pd.read_csv(input_location, index_col=0)
strokes = strokes.applymap(ast.literal_eval)
strokes
```

This data is just a series of x and y coordinates with timestamps (in msec since epoch). As an example we will plot the first stroke.

To make plots we will use matplotlib.

```python
from matplotlib import pyplot as plt
```

```python
def getXVals(df, stroke):
    return (df.iloc[stroke][strokes.iloc[stroke] != 0]).apply(lambda x: x[0])

def getYVals(df, stroke):
    return (df.iloc[stroke][strokes.iloc[stroke] != 0]).apply(lambda x: x[1])
```

```python
plt.title("Strokes example")
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")

for stroke in range(len(strokes)):
    plt.plot(getXVals(strokes, stroke), getYVals(strokes, stroke))

plt.show()
```

It is clear the coordinates need to be flipped over the y-axis.

```python
strokes = strokes.applymap(lambda x: 0 if x == 0 else (x[0], x[1]*-1, x[2]))
```

```python
plt.title("Strokes flipped example")
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")

for stroke in range(len(strokes)):
    plt.plot(getXVals(strokes, stroke), getYVals(strokes, stroke))

plt.show()
```

Now to build one sequence of data we will place all strokes in one sequence and add an extra variable to show when a new stroke starts (1 if true 0 else) and whether a point is part of a pen-up or pen-down stroke (1 for and up stroke and 0 else). While doing this we will ignore 0 values, since these were only necessary to store the strokes in a table.

```python
def get_stroke_direction(stroke):
    """ Returns 1 if stroke direction is up and 0 if stroke direction is down. """
    return int(stroke.iloc[-1][1] > stroke.iloc[0][1])
```

```python
index = 10

plt.title("Stroke flipped example")
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")

plt.plot(getXVals(strokes, index), getYVals(strokes, index))

plt.show()

print(get_stroke_direction(strokes.iloc[index][strokes.iloc[index] != 0]))
```

We will use numpy for normalizing the data.

```python
import numpy as np
```

```python
# rtp raw_touch_points
rtps = []

for i in range(len(strokes)):
    stroke = strokes.iloc[i][strokes.iloc[i] != 0]
    direction = get_stroke_direction(stroke)
    
    for j, point in enumerate(stroke):
        rtps.append((point[0], point[1], point[2], direction, int(j == 0)))
        
rtps = np.array(rtps)
```

```python
def plot_rtps(data, name):
    plt.title(name)
    plt.xlabel("x coordinates")
    plt.ylabel("y coordinates")

    plt.plot(data[:, 0], data[:, 1])

    plt.show()
```

```python
plot_rtps(rtps, "Raw touch points example")
```

Now we need to normalize the data. We will do this so the y values fall in between 0 and 1.

Citation from google paper:

In order to keep the system as flexible as possible
with respect to differences in the writing surface, e.g.
area shape, size, spatial resolution, and sampling rate,
we perform some minimal preprocessing:

- Normalization of xi and yi coordinates, by shifting in x such that x0 = 0, and shifting and scaling the writing area isometrically such that the y coordinate spans the range between 0 and 1. In cases where the bounding box of the writing area is unknown we use a surrogate area 20% larger than the observed range of touch points.
- Equidistant linear resampling along the strokes with Î´ = 0.05, i.e. a line of length 1 will have 20 points.

To do this we first move x0.

```python
x0 = rtps[0][0]
rtps = rtps - [x0, 0, 0, 0, 0]
plot_rtps(rtps, "Raw touch points moved x axis")
```

Next we move the lowest y value to 0.

```python
min_y = np.amin(rtps, axis=0)[1]
rtps = rtps - [0, min_y, 0, 0, 0]
plot_rtps(rtps, "Raw touch points moved y axis")
```

Now we scale all the data points.

```python
max_y = np.amax(rtps, axis=0)[1]
rtps = rtps / [max_y, max_y, 1, 1, 1]
plot_rtps(rtps, "Raw touch points scaled")
```

Lastly we subtract the starting timestamp from all to make a list of relative timestamps.

```python
rtps = rtps - [0, 0, rtps[0][2], 0, 0]
```

```python
rtps = rtps.tolist()
```

Lastly we have to sample the data points.

# Normalization

```python
def sampleLine(p0, p1, delta=0.05):
    l = ((p1[0]-p0[0])**2 + (p0[1]-p1[1])**2)**0.5
    num = int(l/delta)
    
    if num == 0:
        return [p0]
    
    sampled_xs = np.linspace(p0[0], p1[0], num)
    sampled_ys = np.linspace(p0[1], p1[1], num)
    sampled_timestamps = np.linspace(p0[3], p1[3], num)
    
    if p0[3] == 0:
        pen_direction = np.zeros(num)
    else:
        pen_direction = np.ones(num)
    
    start_stroke = np.zeros(num)
    
    if p0[4] == 0:
        start_stroke[0] = 1

    return np.stack((sampled_xs, sampled_ys, sampled_timestamps,pen_direction, start_stroke), axis=1).tolist()
```

```python
resampled_rtps = []

for i, datapoint in enumerate(rtps[:-1]):
    s = sampleLine(rtps[i], rtps[i+1])
    resampled_rtps.extend(s)
```

```python
resampled_rtps = np.array(resampled_rtps)
plot_rtps(resampled_rtps, "Resampled raw touch points")
```

```python
diff_rtps = np.concatenate((np.vstack (([0, 0, 0], np.diff(resampled_rtps, axis=0)[:, :3])), resampled_rtps[:, 3:]), axis=1)
```

```python
total_length = np.sum(np.sum(np.diff(rtps, axis=0)[:, :2]**2, axis=1)**(1/2), axis=0)
print(total_length)
```

Then to scale we take the largest timestamp and divide it by this total length to get the scaling factor.

```python
scaling_factor = rtps[-1][2]/total_length
```

```python
rtps = (np.array(rtps) / [1, 1, scaling_factor, 1, 1]).tolist()
print(rtps[-1][2])
```
