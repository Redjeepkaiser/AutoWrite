```python
import numpy as np
import pickle


padded_features = np.load("../../data/processed_data/bezier_features_personal_padded.npy")

with open("../../data/processed_data/alphabet", "rb") as f:
    alphabet = pickle.load(f)
```

```python
import tensorflow as tf

def split_data(features, train_size=0.9):
    size = len(features)
    indices = np.arange(size)
    np.random.shuffle(indices)
    train_samples = int(size * train_size)
    
    x_train = features[indices[:train_samples]]
    x_valid = features[indices[train_samples:]]
    
    return (
        x_train,
        x_valid
    )

x_train, x_valid = split_data(padded_features)
```

```python
LATENT_SIZE = len(alphabet) + 1

N_BLSTM_LAYERS = 5
N_CELLS = 64

LEARNING_RATE = 10**-4

BATCH_SIZE = 8
N_FEATURES = len(x_train[0][0])
N_TIMESTEPS = len(x_train[0])
N_EPOCHS = 10


checkpoint_path_autoencoder = "../../training/checkpoints/autoencoder/cp.ckpt"
checkpoint_path_encoder = "../../training/checkpoints/encoder_bezier_personal/cp.ckpt"
```

```python
def create_dataset(train_data, batch_size):
    dataset = tf.data.Dataset.from_tensor_slices((x_train, x_train))
    dataset = dataset.shuffle(1000)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(1)
    return dataset

train_dataset = create_dataset(x_train, BATCH_SIZE)
validation_dataset = create_dataset(x_valid, BATCH_SIZE)
```

```python
from keras.layers import Bidirectional, LSTM, Dense, Input, Masking

def build_encoder_model(n_blstm_layers, n_cells, n_features, output_size):
    model = tf.keras.models.Sequential()

    model.add(Masking(input_shape=(None, N_FEATURES), mask_value=np.zeros((N_FEATURES))))
    
    for i in range(n_blstm_layers):
        model.add(Bidirectional(LSTM(N_CELLS,
                                     input_shape=(None, N_FEATURES),
                                     return_sequences = True,
                                     dropout = 0.5),
                                merge_mode = 'sum'))

    model.add(Dense(output_size, activation = 'softmax'))
    return model

def build_decoder_model(n_blstm_layers, n_cells, n_features, output_size):
    model = tf.keras.models.Sequential()
    
    for i in range(n_blstm_layers):
        model.add(Bidirectional(LSTM(N_CELLS,
                                     input_shape=(None, N_FEATURES),
                                     return_sequences = True,
                                     dropout = 0.5),
                                merge_mode = 'sum'))

    model.add(Dense(output_size))
    return model
```

```python
from tensorflow.keras.models import Model
import os

class Autoencoder(Model):
    def __init__(self, latent_dim, checkpoint_path):
        super(Autoencoder, self).__init__()
        self.latent_dim = latent_dim
        
        self.encoder = build_encoder_model(N_BLSTM_LAYERS, N_CELLS, N_FEATURES, LATENT_SIZE)
        
        self.encoder.load_weights(checkpoint_path)
        self.decoder = build_decoder_model(N_BLSTM_LAYERS, N_CELLS, 1, N_FEATURES)

    def call(self, x):
        encoded = self.encoder(x)
        print(encoded)
        reshaped_output = tf.reshape(encoded, (encoded.shape[1], encoded.shape[0], encoded.shape[2]))
        (decoded, log_probabilities) = tf.nn.ctc_beam_search_decoder(reshaped_output, [reshaped_output.shape[0]], beam_width=3)
        decoded = self.decoder(decoded)
        return decoded

autoencoder = Autoencoder(len(alphabet), checkpoint_path_encoder)
```

```python
from tensorflow.keras import losses

autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())
```

```python
from pathlib import Path
import pickle

loss_history = [[], []]

loss_path = Path("../../training/logs/autoencoder/loss")
if loss_path.is_file():
    with open(loss_path, "rb") as f:
        loss_history = pickle.load(f)
```

```python
import keras

# Create a callback that saves the model's weights
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path_autoencoder,
                                                 save_weights_only=True,
                                                 verbose=1)

history = autoencoder.fit(train_dataset,
                epochs=N_EPOCHS,
                validation_data=validation_dataset,
                callbacks=[cp_callback]
)

loss_history[0].extend(history.history['loss'])
loss_history[1].extend(history.history['val_loss'])

with open(loss_path, "wb") as f:
    pickle.dump(loss_history, f)
```

```python
import matplotlib.pyplot as plt

def flatten_list(l):
     return [point for elem in l for point in elem]

plt.plot(loss_history[0], color="red", label="train")
plt.plot(loss_history[1], color="black", label="test")
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.show()
```
